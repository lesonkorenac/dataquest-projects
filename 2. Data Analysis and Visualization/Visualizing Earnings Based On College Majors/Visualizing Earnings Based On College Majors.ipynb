```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pandas.tools.plotting import scatter_matrix
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
print("All data count: %s" % recent_grads.shape[0])

recent_grads.dropna(inplace=True)
print("Cleaned data count: %s" % recent_grads.shape[0])
```

```python
def relation_scatter_plot(x_key, y_key):
    recent_grads.plot(x=x_key, y=y_key, kind="scatter")
    plt.title("[{} - {}] relation".format(x_key, y_key))
    plt.show()
    
ralations =[
    ("Sample_size", "Median"),
    ("Sample_size", "Unemployment_rate"),
    ("Full_time", "Median"),
    ("ShareWomen", "Unemployment_rate"),
    ("Men", "Median"),
    ("Women", "Median")
]    
    
for relation in ralations:
    relation_scatter_plot(relation[0], relation[1])
```
